```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flower_df=sns.load_dataset("iris")
```

```python
flower_df
```

```python
flower_df.species.unique()
```

```python
plt.figure(figsize=(12,6))
plt.title("Sepal Dimention")
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width,hue=flower_df.species,s=50);
```

```python
setosa_df=flower_df[flower_df.species=="setosa"]
versicolor_df=flower_df[flower_df.species=="versicolor"]
virginica_df=flower_df[flower_df.species=="virginica"]
```

```python
sns.set_style("darkgrid")
```

```python

plt.title("Sepal Width of Different Species")
plt.xlabel("width")
plt.ylabel("No.of flowers")
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True)
plt.legend(["setosa","versicolor","virginica"]);
```

```python
setosaP_df=flower_df[flower_df.species=='setosa']
versicolorP_df=flower_df[flower_df.species=='versicolor']
virginicaP_df=flower_df[flower_df.species=='virginica']
```

```python
plt.title("Distribution of Length")
plt.hist([setosaP_df.petal_length,versicolorP_df.petal_length,virginicaP_df.petal_length],stacked=True)
plt.xlabel('length')
plt.ylabel('No.of Flowers');
```
