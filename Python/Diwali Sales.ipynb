```python
Diwali_df
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=Diwali_df["Gender"],edgecolor="Black",width=0.4,hue=Diwali_df["Gender"],palette="deep")
plt.title("Gender Count");
```

```python
Amt=Diwali_df.groupby("Gender")["Amount"].sum()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(Amt,edgecolor="black",width=0.4,palette="PuOr",saturation=0.9)
plt.title("Amount By Gender")
;
```

```python
Age_grps=Diwali_df.groupby("Age Group")["Gender"].count()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(Age_grps,palette="Greens",edgecolor="black")
plt.title("Different Age Groups");
```

```python
plt.figure(figsize=(10,6))
ax_label=sns.countplot(x="Age Group",data=Diwali_df,hue="Gender",palette="Purples",edgecolor="black")
for bars in ax_label.containers:
    ax_label.bar_label(bars)
```

```python
Total_amt_by_age=Diwali_df.groupby("Age Group")["Amount"].sum()
Total_amt_by_age
```

```python
plt.figure(figsize=(10,6))
sns.barplot(Total_amt_by_age,palette="Purples",edgecolor="black");
```

```python
count_States=Diwali_df["State"].value_counts()
count_States
```

```python
plt.figure(figsize=(10,6))
sns.barplot(count_States,palette="Reds",edgecolor="Black")
plt.xticks(rotation=75)
plt.title("State wise Sales");
```

```python
count_Ma=Diwali_df["Marital_Status"].value_counts()
count_Ma
```
